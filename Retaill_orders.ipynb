```python
#read data from the file and handle null values
import pandas as pd
df= pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df.head(10)
df['Ship Mode'].unique()
```

```python
df.info()
```

```python
df.head(10)
```

```python
#rename the columns to lower case and add underscore between 2 words
#df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(' ','_')
```

```python
df.head()
```

```python
#derive new columns discount, sales price and profit
#df['discount']= df['list_price']*df['discount_percent']*.01
#df['sale_price']= df['list_price'] - df['discount']
df['profit']= df['sale_price']- df['cost_price']
df.head()
```

```python
# change the order date datatype to date
df['order_date']= pd.to_datetime(df['order_date'], format= "%Y-%m-%d")
df.head()
```

```python
df.info()
```

```python
#drop columns like cost_price, list_price, discount_percent  
df.drop(['cost_price', 'list_price', 'discount_percent'], axis= 1, inplace= True)
df
```

### Import needed libraries

```python
import pandas as pd
from sqlalchemy import create_engine
from pandas.io import sql
```

### create engine to connect sql and python

```python
engine= create_engine("mysql+pymysql://{user}:{pw}@localhost:3306/{db}"
                      .format(user="root",
                              pw="***",
                              db="orders_database"))
```

### move the dataframe to database i.e(python to sql)

```python
df.to_sql(con=engine,name="orders",if_exists="append", index= False)
```
